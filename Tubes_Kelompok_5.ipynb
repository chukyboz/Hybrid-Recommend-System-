```python
from surprise import Dataset, Reader, SVD, KNNBasic, BaselineOnly
from surprise.model_selection import train_test_split, cross_validate
from surprise import accuracy
from collections import defaultdict
import pandas as pd

# Assuming 'final_df' is already defined and has the correct format
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(final_df[['RecipeId', 'AuthorId', 'Rating']], reader)

# Define the algorithms
algorithms = {
    'SVD': SVD(),
    'KNNBasic': KNNBasic(),
    'BaselineOnly': BaselineOnly()
}

# Split data into training and test set
trainset, testset = train_test_split(data, test_size=0.25)

# Function to get top-N recommendations for each user
def get_top_n(predictions, n=10):
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]
    return top_n

# Train and evaluate the algorithms
performance_metrics = {}
top_n_recommendations = {}
for name, algo in algorithms.items():
    # Train the algorithm on the trainset
    algo.fit(trainset)
    # Test on the testset
    predictions = algo.test(testset)
    # Get top-N recommendations
    top_n = get_top_n(predictions, n=10)
    # Convert the top N recommendations into a DataFrame
    top_n_df = pd.DataFrame([(uid, iid, est) for uid, user_ratings in top_n.items() for (iid, est) in user_ratings],
                            columns=['AuthorId', 'RecipeId', 'EstimatedRating']).sort_values(by=['EstimatedRating'], ascending=False)
    # Keep only the top 10 recommendations
    top_n_recommendations[name] = top_n_df.head(10)

# Output the results
for name in algorithms.keys():
    print(f"Algorithm: {name}")
    print(f"Top 10 Recommendations:\n{top_n_recommendations[name]}")

```

```python
# Assuming 'final_df' is already defined and has the correct format
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(final_df[['RecipeId', 'AuthorId', 'Rating']], reader)

# Define the algorithms
algorithms = {
    'SVD': SVD(),
    'KNNBasic': KNNBasic(),
    'BaselineOnly': BaselineOnly()
}

# Split data into training and test set
trainset, testset = train_test_split(data, test_size=0.25)

# Function to get top-N recommendations for each user
def get_top_n(predictions, n=10):
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]
    return top_n

# Train and evaluate the algorithms
performance_metrics = {}
all_recommendations = []
for name, algo in algorithms.items():
    # Train the algorithm on the trainset
    algo.fit(trainset)
    # Test on the testset
    predictions = algo.test(testset)
    # Calculate RMSE and MAE
    rmse_value = accuracy.rmse(predictions)
    mae_value = accuracy.mae(predictions)
    performance_metrics[name] = {'RMSE': rmse_value, 'MAE': mae_value}
    # Get top-N recommendations
    top_n = get_top_n(predictions, n=10)
    # Convert the top N recommendations into a DataFrame
    top_n_df = pd.DataFrame([(uid, iid, est) for uid, user_ratings in top_n.items() for (iid, est) in user_ratings],
                            columns=['AuthorId', 'RecipeId', 'EstimatedRating'])
    # Add algorithm name to the DataFrame
    top_n_df['Algorithm'] = name
    # Append to the list of all recommendations
    all_recommendations.append(top_n_df)

# Combine all recommendations into a single DataFrame
combined_recommendations = pd.concat(all_recommendations)

# Group by RecipeId and AuthorId to average the ratings from different algorithms
grouped = combined_recommendations.groupby(['RecipeId', 'AuthorId'])
averaged_recommendations = grouped['EstimatedRating'].mean().reset_index()

# Sort the averaged recommendations to get the overall top 10
overall_top_10_df = averaged_recommendations.sort_values(by='EstimatedRating', ascending=False).head(10)

# Calculate the average RMSE and MAE across all algorithms
average_rmse = sum(metric['RMSE'] for metric in performance_metrics.values()) / len(algorithms)
average_mae = sum(metric['MAE'] for metric in performance_metrics.values()) / len(algorithms)

# Output the results
print("Overall Top 10 Combined Recommendations from All Algorithms:")
print(overall_top_10_df)
print(f"\nAverage RMSE: {average_rmse}")
print(f"Average MAE: {average_mae}")
```

CONTENT BASED FILTERING

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# final_dfr = ...

# Menggabungkan kolom teks untuk membuat deskripsi lengkap
final_dfr['combined_features'] = final_dfr[['Description', 'Keywords', 'RecipeIngredientParts']].fillna('').agg(' '.join, axis=1)

# Inisialisasi TF-IDF Vectorizer
tfidf = TfidfVectorizer(stop_words='english')

# Menghitung matriks TF-IDF
tfidf_matrix = tfidf.fit_transform(final_dfr['combined_features'])

# Menghitung skor kesamaan cosine
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Mengonversi skor kesamaan cosine ke DataFrame
cosine_sim_df = pd.DataFrame(cosine_sim, index=final_dfr['RecipeId'])

# Tampilkan output
print(cosine_sim_df)
```

```python

# Function to get similar recipes with separate columns for AuthorId and RecipeId
def get_similar_recipes(recipe_id, top_n=10):
    # Get the index of the recipe from the given ID
    idx = final_dfr.index[final_dfr['RecipeId'] == recipe_id].tolist()[0]

    # Get similarity scores for this recipe with all other recipes
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort recipes based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the top recipes' IDs and AuthorIds
    top_recipe_ids = [final_dfr['RecipeId'].iloc[i[0]] for i in sim_scores[1: top_n+1]]
    top_author_ids = [final_dfr['AuthorId'].iloc[i[0]] for i in sim_scores[1: top_n+1]]

    return pd.DataFrame({'RecipeId': top_recipe_ids, 'AuthorId': top_author_ids})

# Example usage
similar_recipes_dfr = get_similar_recipes(487)
print("top 10 recommendation")
print(similar_recipes_dfr)
```
